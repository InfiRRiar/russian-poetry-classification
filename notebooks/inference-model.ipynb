```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
```

```python
label_to_author = {
    0: "blok",
    1: "cvetaeva",
    2: "pasternak"
}
```

```python
trained_model_path = "../models/alpha"
model_name = "cointegrated/rubert-tiny2"

device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
model = AutoModelForSequenceClassification.from_pretrained(trained_model_path, num_labels=3)
tokenizer = AutoTokenizer.from_pretrained(model_name)

model.eval()
```

```python
x = """Весь мир не дышал, пока мы стояли
У звёздного неба, обнявшись вдвоём.
Открыты сердца, что не понимали,
Как завтрашний день мы с тобой проживём.

Глубокая ночь, но тонут всё дольше
В глазах у друг друга, сверкая, глаза.
Искр меж нами становится больше -
Часть наша, а часть принесли небеса.

Над нами пожар, гадать только можно,
Действительно ль звёзды в округе - не сон?
Когда наяву это стало возможно?
Сердца бьются чаще, всю жизнь в унисон...

Сплелись наши судьбы в единые нити,
Их видит весь мир, ведь в небе горя,
Дают они свет - неспящих спаситель,
Не сгубят их тени, но сгубит заря.

Мы знаем, что скоро наступит мгновенье,
Как солнце взойдёт и закончится бал.
Пока что идёт мёртвых звёзд представленье,
А чувств до него я как будто не знал...

Никто нас не ждёт, никто нас не ищет,
Но день начинать надо завтра с утра;
Диктует устав столь обыденной жизни - 
По разным путям расходиться пора.

Проснёмся не вместе, как, впрочем, обычно,
Минуты под солнцем - как будто бы ад.
День выглядит странно и так непривычно,
Когда в наших жизнях идёт звездопад..."""
```

```python
tokenized_x = tokenizer(x, return_tensors="pt", truncation=False, padding=True)

predict = model(**tokenized_x)

probs = torch.softmax(predict.logits, dim=1)
for k, v in enumerate(probs.flatten()):
    print(f"{label_to_author[k]}: {v}")

predict = torch.argmax(predict.logits, dim=1)

print(label_to_author[predict.item()])
```
